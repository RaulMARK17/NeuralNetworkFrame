```python
import numpy as np
import NNframework as NF
```

```python
X = np.float32(np.array([
    [0,0],
    [1,0],
    [0,1],
    [1,1]]))
Y = np.float32(np.array([
    [1,1],
    [1,0],
    [0,1],
    [0,0]]))
```

```python
NN = NF.NeuralNetwork()
NN.createNetwork()
```

```python
NN.visualNN()
```

```python
NN.train(X, Y, .8, .001, 10000, period = 20)
```

```python
NN.predict(X)
```

```python
NN.saveNetwork('NN/')
```

```python
NN2 = NF.NeuralNetwork()
```

```python
NN2.loadNetwork('NN/') 
```

```python
NN2.predict([1, 1])
```

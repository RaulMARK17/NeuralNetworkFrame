## Si viste "Game Of Thrones" sabrás que en la batalla de los bastardos nuestros heroes quedan atrapados, entonces, los de azul son los malos y la red neuronal debe aprender a seleccionar unicamente a los que están de rojo.

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
import NNframework as NF
```

```python
# DATASET
n = 500

X, Y = make_circles(n_samples = n, factor = .5, noise=.05)

Y = Y[:, np.newaxis]

plt.scatter(X[Y[:, 0] == 0, 0], X[Y[:, 0] == 0, 1], c = 'skyblue')
plt.scatter(X[Y[:, 0] == 1, 0], X[Y[:, 0] == 1, 1], c = 'salmon')
plt.axis('equal')
plt.show()


#Basicamente, este es el dataset.
```

```python
nn = NF.NeuralNetwork() # Se crea la red.
nn.createNetwork([2, 4, 8, 4, 1], ['relu', 'sigmoid', 'sigmoid', 'sigmoid']) # Se pasan los parametros de la estructura. 
```

```python
nn.visualNN() 
```

# El codigo siguiente es obtenido del ejemplo de Dot CSV y solo es para una demostración más gráfica al momento del entrenamiento.

## El proceso de entrenamiento se desgloza un poco, pero sigue siendo con el código base del framework. 

```python
import time
from IPython.display import clear_output

loss = []
lr = 0.05
epochs = 2500
for i in range(epochs):
  # Entrenemos a la red!
  pY = nn.predict(X)
  nn.backPropagation(X, Y, lr, pY)
  if i % 20 == 0:
    if lr > 0.001:
      lr -= 0.001
  
  if i % (epochs/100) == 0:     
    loss.append(nn.e(Y, pY, deriv=False))

    res = 50

    _x0 = np.linspace(-1.5, 1.5, res)
    _x1 = np.linspace(-1.5, 1.5, res)

    _Y = np.zeros((res, res))

    for i0, x0 in enumerate(_x0):
      for i1, x1 in enumerate(_x1):
        _Y[i0, i1] = nn.predict(np.array([[x0, x1]]))

    plt.pcolormesh(_x0, _x1, _Y, cmap="magma")
    plt.axis("equal")
    plt.scatter(X[Y[:,0] == 0, 0], X[Y[:,0] == 0, 1], c="skyblue")
    plt.scatter(X[Y[:,0] == 1, 0], X[Y[:,0] == 1, 1], c="salmon")
    plt.show()
    print(i)
    clear_output(wait=True)
    time.sleep(0.1)

plt.pcolormesh(_x0, _x1, _Y, cmap="magma")
plt.axis("equal")
plt.scatter(X[Y[:,0] == 0, 0], X[Y[:,0] == 0, 1], c="skyblue")
plt.scatter(X[Y[:,0] == 1, 0], X[Y[:,0] == 1, 1], c="salmon")
plt.show()
nn.plot(loss)
```
